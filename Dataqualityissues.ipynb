```python
import json
import pandas as pd
```

```python
with open('users.json') as f:
  data = json.load(f)
```

```python
print(json.dumps(data, indent = 4, sort_keys=True)) #have used 4 spaces for indentation. And, the keys are sorted in ascending order
```

```python
#Flattening nested list from JSON object
df_users = pd.json_normalize(data,meta=[
        ['_id','$oid'], 'active',
        ['createdDate', '$date'], 
        ['lastLogin', '$date'], 'role','signUpSource', 'state'])
```

```python
print(df_users)
```

```python
brands = []
for line in open('brands.json', 'r'):
    brands.append(json.loads(line))
```

```python
#Flattening nested list from JSON object
df_brands = pd.json_normalize(brands,meta=[
        ['_id','$oid'], 'barcode','category', 'categoryCode',
        ['cpg', '$id','$oid'], 
        ['cpg', '$ref'], 'name','topBrand'])
```

```python
print(df_brands.head())
```
